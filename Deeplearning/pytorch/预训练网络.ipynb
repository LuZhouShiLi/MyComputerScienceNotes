```python
from torchvision import models

dir(models)
```

```python
alexnet = models.AlexNet()  ## 创建alexnet实例

resnet = models.resnet101(pretrained = True)   # 实例化101层的卷积神经网络
```

```python
resnet
```

```python
from torchvision import transforms

## 定义基本与处理函数的管道
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean = [0.485,0.456,0.406],
        std = [0.229,0.224,0.225]
)])
```

```python
from PIL import Image
img =  Image.open("D:\\Pictures\\Talyor\\1.jpg")  ## 使用Pillow加载图像
img.show()

img_t = preprocess(img)  ## 使用预处理管道传递图像

import torch
batch_t =   torch.unsqueeze(img_t,0)  # 在指定索引插入一个维度

```

```python
resnet.eval()  ## 设置网络为推理模式  就是验证模式

out = resnet(batch_t)
out

```

```python
## 加载1000个标签的文件
with open('./dlwpt-code/data/p1ch2/imagenet_classes.txt') as f:
    ## 删除空格和换行符
    labels = [line.strip() for line in f.readlines()]

## 确定out张量中最高分对应的索引

_,index = torch.max(out,1)

## 使用索引来访问标签   输出规划到0~1 
percentage = torch.nn.functional.softmax(out,dim = 1)[0] * 100
labels[index[0]],percentage[index[0]].item()


_,indices = torch.sort(out,descending=True)
[(labels[idx],percentage[idx].item()) for idx in indices[0][:5]]

```
